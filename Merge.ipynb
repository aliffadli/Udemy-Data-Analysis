```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
```

```python
#jadinya key ada dalam bentuk kolom
#pertanyaan untuk dijawab: kira-kira bisa tidak key dan data_set_1 jadinya berbentuk baris?
dframe1 = DataFrame({'key':['X', 'Z','Y','Z','X','X'], 'data_set_1':np.arange(6)})
dframe1
```

```python
#jawaban untuk pertanyaan di atas: bisa, dengan menggunakan method .transpose()
dframe1.transpose()
```

```python
dframe2 = DataFrame({'key':['Q','Y','Z'], 'data_set_2':[1,2,3]})
dframe2
```

```python
#melihat 'key' yang sama dari kedua dframe tersebut, kalau yang cuma ada satu, maka tidak dimasukkan ke dalam dframe yang baru
#kenapa yang dilihat key? kenapa bukan data_set_1 atau data_set_2? mungkin karena 'key' adalah nama key yang ada di kedua dframe
pd.merge(dframe1, dframe2)
```

```python
#menggunakan join juga tidak bisa, tidak mengerti kenapa 
#dframe1.join(dframe2)
```

```python
#on sebenarnya digunakan untuk memilih kolom secara spesifik
#cuma dalam kasus ini, 'on' menggunakan key yang sama
pd.merge(dframe1, dframe2, on='key')
```

```python
#kalau pakai cara seperti di bawah akan muncul KeyError
#pd.merge(dframe1, dframe2, on='data_set_1')
```

```python
dframe3 =  DataFrame({'key':['B','Y','Z'], 'data_set_3':[2,2,3]})
dframe3
```

```python
#tidak bisa melakukan merge lebih dari tiga dalam satu method, akan muncul error: Value Error
#pd.merge(dframe1, dframe2, dframe3)

dfmerge1 = pd.merge(dframe1, dframe2)
dfmerge1
```

```python
#baru bisa ketika dipanggil .merge dua kali
#agak sulit dan lama (?) atau best practice? 
dfmerge2 = pd.merge(dfmerge1, dframe3)
dfmerge2
```

```python
#acuannya adalah key yang ada di dframe1
pd.merge(dframe1, dframe2, on='key', how='left')
#.merge hampir mirip dengan .join()
```

```python
#acuannya adalah key yang ada di dframe2
#kenapa Z-nya ada dua? karena Z di sebelah kiri (dframe1) punya dua value
#perhatikan di dframe2 walaupun ada dua Z tapi valuenya tidak berubah
#setiap key yang sama maka dibuat berulang, apakah 2x1 (?) mari kita coba
pd.merge(dframe1, dframe2, on='key', how='right')
```

```python
dframe4 = DataFrame({'key':['Z','Z','Z','Q','A'], 'data_set_3':[1,2,4,4,3]})
dframe4
```

```python
dframe5 = DataFrame({'key':['Y','V','Z','Z'], 'data_set_4':[1,7,2,4]})
dframe5
```

```python
pd.merge(dframe4,dframe5,on='key',how='left')
```

```python
pd.merge(dframe4,dframe5,on='key',how='right')
```

```python
#rumusnya adalah baris kiri dikali baris kanan
```

```python
#jika kita mau menggunakan baik yang kiri maupun yang kanan, bisa dengan menggunakan outer, maka semua key akan masuk
pd.merge(dframe1, dframe2, on='key', how='outer')
```

```python
#inner juga bisa, inner hanya untuk key yang dimiliki oleh kedua dataset
#untuk membuat inner, kita tidak perlu membuat parameter how sebenarnya, karena .merge sudah automatis membuat .merge
pd.merge(dframe1, dframe2, on='key', how='inner')
```

```python
df_left = DataFrame({'key1':['SF','SF','LA'], 
                    'key2': ['one','two','one'],
                    'left_data':[10,20,30]})
```

```python
df_right = DataFrame({'key1':['SF','SF','LA','LA'],
                     'key2':['one','one','one','two'],
                     'right_data':[40,50,60,70]})
```

```python
df_left
```

```python
df_right
```

```python
#key1 dan key2 merupakan sumbunya atau pusatnya
#bagian .merge ini masih belum begitu mengerti 
#bagian ini agak challanging dan perlu banyak latihan
pd.merge(df_left,df_right, on=['key1','key2'], how='outer')
```

```python
#jika kita cuma mau merge hanya terhadap key1, bisa. Namun, jika kita memiliki dua kolom yang sama, maka kolom tersebut akan diganti juga namanya
#seperti yang tertera di bawah, key2 jadi key2_x untuk dframe kiri dan key2_y untuk dframe kanan
pd.merge(df_left, df_right, on='key1')
```

```python
#jika kita mau mengubah akhiran dari '_x' atau '_y' tersebut
pd.merge(df_left, df_right, on='key1', suffixes=('_lefty', '_righty'))
#maka _x dan _y tersebut berubah menjadi lefty dan righty 
```

```python
#ini yang inner
pd.merge(df_left, df_right)
```
