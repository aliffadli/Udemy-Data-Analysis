## Mulai sekarang akan menggunakan nfl_frame2

```python
#tidak ada stadium, karena memang di kolom tidak ada 
DataFrame(nfl_frame2,columns=['Rank', 'Team', 'Stadium'])
```

```python
#memasukkan value ke dalam kolom stadium
nfl_frame2['Stadium'] = "Levi's Stadium"

#semua stadium mempunyai value Levi's Stadium, karena cuma memasukkan value Levi's Stadium
nfl_frame2
```

```python
#perhatikan bahwa np.arange artinya memanggil array dari 0-10 dengan index juga sama
#array tersebut dimasukkan ke dalam DataFrame di kolom Stadium

#kesimpulan: jika kami memasukkan array sepanjang 10 elemen ke dalam kolom apapun (tidak hanya stadium), maka valuenya juga berubah
nfl_frame2['Stadium'] = np.arange(10)
nfl_frame2
```

```python
#mengambil sampel lebih kecil karena ukuran 10 terlalu banyak 
frame_nfl = nfl_frame2.copy().iloc[0:6]
frame_nfl
```

```python
#memasukkan series ke dalam dataframe

#fun fact: kutipnya bisa beda untuk per elemen
stadiums = Series(["Levis's Stadium", 'AT&T Stadium'], index=[4,0])
stadiums
```

```python
frame_nfl['Stadium'] = stadiums
frame_nfl
```

```python
#mengubah jumlah menang dengan menggunakan slicing, tapi mendapat warning
#asumsi saya adalah karena yang saya lakukan adalah slicing dan bukan copy. jadi hal itu berbeda
#mungkin bisa dibilang gak best practice (?)
frame_nfl['Won'].iloc[4] = 23
```

```python
frame_nfl
```

```python
#syntax bisa 12 saja tanpa kurung siku karena elemen yang dimasukkan tunggal
tied = Series(12, index=[1])
tied
```

```python
#menarik. awalnya saya kira tied, cuma mengubah satu elemen saja yaitu yang dipilih, ternyata mengubah semua elemen
#semua kolom Tied menjadi NaN karena variabel tied yang didalamnya cuma ada value 12 dan index 1

#REVISITED: Define Series cuma bisa memasukkan nilai, tidak bisa memasukkan kolom
#memasukkan kolom menggunakan command lain, bukan Series(___)
frame_nfl['Tied'] = tied
frame_nfl

#jadi solusi paling baik untuk mengubah spesifik elemen adalah menggunakan iloc/loc? 
```

```python
#menghapus kolom 
del frame_nfl['Tied']
```
