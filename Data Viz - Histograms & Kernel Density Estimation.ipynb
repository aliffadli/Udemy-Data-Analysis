```python
#kegunaan dari kernel density estimation adalah mewakilkan tiap tickmark dengan gaussian basis function
sns.rugplot(dataset3)

x_min=dataset3.min() -2
x_max=dataset3.max() +2

x_axis=np.linspace(x_min,x_max,100)

bandwidth=((4*dataset3.std()**5)/(3*len(dataset3))) ** 0.2

kernel_list= []

for data_point in dataset3:
    #Create a kernel in each point and append it to the kernel_list
    kernel = stats.norm(data_point, bandwidth).pdf(x_axis)
    kernel_list.append(kernel)
    
    #Scale for plotting
    kernel = kernel / kernel.max()
    kernel = kernel * 0.4
    
    plt.plot(x_axis, kernel, color='grey', alpha=0.5)
    
#menampilkan gaussian dari grafik
#saya juga belum begitu mengerti apa yang dimaksud dengan gaussian
plt.ylim(0,1)
```

```python
sum_of_kde=np.sum(kernel_list, axis=0)
fig = plt.plot(x_axis,sum_of_kde,color='indianred')
sns.rugplot(dataset3)
plt.yticks([])
plt.suptitle("Sum of the basis function")
```

```python
#menggunakan seaborn bisa lebih simpel
sns.kdeplot(dataset3)
```

```python
sns.rugplot(dataset3,color='black')

#specified particular bandwith
for bw in np.arange(0.5,2,0.25):
    sns.kdeplot(dataset3,bw=bw,lw=1.8,label=bw)
```

```python
#jika ingin lebih mengetahui mengenai kernel
url = 'http://en.wikipedia.org/wiki/Kernel_(statistics)'
```

```python
# MULTIVARIATE DENSITY ESTIMATION 

#Multivariate artinya melihat hubungan antara dua variabel dengan menggunakan kde 

# Mean center of data
mean = [0,0]

# Diagonal covariance
cov = [[1,0],[0,100]]

# Create dataset using numpy
dataset2 = np.random.multivariate_normal(mean,cov,1000)

# Bring back our old friend pandas
dframe = pd.DataFrame(dataset2,columns=['X','Y'])

# Plot our dataframe
sns.kdeplot(dframe)
```

```python
# We could have also passed two vectors seperately, and shade
sns.kdeplot(dframe.X,dframe.Y,shade=True)
```

```python
# Can specify a particualr bandwidth
sns.kdeplot(dframe,bw=1)
```

```python
# Or just use silverman again
sns.kdeplot(dframe,bw='silverman')
```

```python
# We can also create a kde joint plot, simliar to the hexbin plots we saw before

sns.jointplot('X','Y',dframe,kind='kde')
```
