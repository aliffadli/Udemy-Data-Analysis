## JSON with Python

```python
json_obj = """
{   "zoo_animal": "Lion",
    "food": ["Meat", "Veggies", "Honey"],
    "fur": "Golden",
    "clothes": null, 
    "diet": [{"zoo_animal": "Gazelle", "food":"grass", "fur": "Brown"}]
}
"""
```

```python
import json
```

```python
#load dari json data, tapi masih belum diketahui kegunaannya untuk apa
data = json.loads(json_obj)
```

```python
data
```

```python
#covert balik menjadi objek json
json.dumps(data)
```
