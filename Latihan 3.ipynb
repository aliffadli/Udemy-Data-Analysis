```python
feature_cols = ['Life_Expectancy', 'Adult_Mortality', 'Infant_Deaths', 'Alcohol', 'Percentage_Exp', 'HepatitisB', 'Measles', 'BMI',
                'Under_Five_Deaths', 'Polio', 'Tot_Exp', 'Diphtheria', 'HIV/AIDS', 'GDP', 'Population', 'thinness_1to19_years', 
                'thinness_5to9_years', 'Income_Comp_Of_Resources', 'Schooling']
X2 = le[feature_cols]
y2 = le['Status']
```

```python
from sklearn import preprocessing

#Melakukan normalisasi dengan z-score karena banyak outliers
standard_scale = preprocessing.StandardScaler()
X2 = standard_scale.fit_transform(X2)
```

```python
from sklearn.model_selection import train_test_split

X2_train, X2_test, y2_train, y2_test = train_test_split(X2, y2, test_size=0.33)
```

```python
from sklearn.neighbors import KNeighborsClassifier

#Initialize our classifier
KKN = KNeighborsClassifier(n_neighbors = 5)

#Train our model
KKN.fit(X2_train, y2_train)

#make prediction using test data
preds = KKN.predict(X2_test)
print(preds)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score

print('akurasi: ', accuracy_score(y2_test, preds))
print('recall: ', recall_score(y2_test, preds))
print('precision: ', precision_score(y2_test, preds))
```

```python
import itertools
from sklearn.metrics import confusion_matrix

def plot_confusion_matrix(cm, classes, normalize=False, title='Confusion matrix', cmap=plt.cm.Blues) :
    
    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:,np.newaxis]
        print("Normalized confusion matrix")
    else:
        print("Confusion matrix, without normalization")
    print(cm)
    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)
    fmt = '.2f' if normalize else 'd'
    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, format(cm[i, j], fmt),
                horizontalalignment="center",
                color="white" if cm[i, j] > thresh else "black")
        plt.tight_layout()
        plt.ylabel('True label')
        plt.xlabel('Predicted label')
```

```python
#Plot confusion matrix
conf_mat = confusion_matrix(y2_test, preds)
plt.figure()
plot_confusion_matrix(conf_mat, classes=['developed', 'developing'],
                     normalize=False, title='Confusion Matrix')
plt.show()
```
