### SVM with Sci Kit Learn

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

Import iris dataset

```python
from sklearn import datasets

iris = datasets.load_iris()

#Grab feature X and target Y 
X = iris.data

Y = iris.target

print(iris.DESCR)
```

```python
#import SVC (Support Vector Classification)
from sklearn.svm import SVC
```

```python
model = SVC()
```

```python
#split data 
from sklearn.model_selection import train_test_split 

X_train, X_test, Y_train, Y_test = train_test_split(X,Y)
```

```python
model.fit(X_train, Y_train)
```

```python
from sklearn import metrics

#Get predictions
predicted = model.predict(X_test)
expected = Y_test

print(metrics.accuracy_score(expected, predicted))
```

Looks like we have achieved a 97.4 % accuracy with Support Vector Classification!

Now that we've gone through a basic implementation of SVM lets go ahead and quickly explore the various kernel types we can use for classification. We can do this by plotting out the boundaries created by each kernel type! We'll start with some imports and by setting up the data.

The linear models LinearSVC() and SVC(kernel='linear') yield slightly different decision boundaries. This can be a consequence of the following differences:

- LinearSVC minimizes the squared hinge loss while SVC minimizes the regular hinge loss.
- LinearSVC uses the One-vs-All (also known as One-vs-Rest) multiclass reduction while SVC uses the One-vs-One multiclass reduction.

NOTE: kernel biasa digunakan untuk mengkalkulasikan pattern yang tidak linear. Seperti dijelaskan dalam dokumentasi nbviewer, kernel digunakan untuk slicing data atau memotong data tapi tidak linear (berupa garis lurus).

```python
#import all SVM
from sklearn import svm

#X cuma mengambil dua kolom perta
X = iris.data[:,:2]
Y = iris.target

#SVM regularization parameter 
#Hal ini digunakan agar setiap kernel method memiliki regularization yang sama 
#Agar kita bisa melakukan komparasi yang akurat
C = 1.0

#SVC with linear kernel
svc = svm.SVC(kernel='linear', C=C).fit(X,Y)

# Gaussian Radial Bassis Function
rbf_svc = svm.SVC(kernel='rbf', gamma=0.7, C=C).fit(X,Y)

#SVC with 3rd degree polynomial 
poly_svc = svm.SVC(kernel='poly', degree=3, C=C).fit(X,Y)

# SCV linear
lin_svc = svm.LinearSVC(C=C).fit(X,Y)
```

#### Visualize the plot

First we define a mesh to plot in. We define the max and min of the plot for the y and x axis by the smallest and larget features in the data set. We can use numpy's built in mesh grid method to construct our plot.

```python
#Set the step size (tidak begitu mengerti kegunaan dari step size)
h = 0.02

#X axis min and max
x_min = X[:,0].min() - 1
x_max = X[:,0].max() + 1

#Y axis min and max
y_min = X[:,1].min() - 1
y_max = X[:,1].max() + 1

# Create meshgrid 
xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))
```

```python
#title for the plots 
titles = ['SVC with linear kernel', 'LinearSVC (linear kernel)', 'SVC with RBF kernel', 'SVC with polynomial (degree 3 kernel)']
```

Finally we will go through each model, set its position as a subplot, then scatter the data points and draw a countour of the decision boundaries.

```python
#use enumerate for a count
# i merupakan iterasi ke berapa, sementara clf merupakan objek yang diiterasikan (apakah sedang iterasi svc, lin_svc, yang lain)
for i, clf in enumerate((svc, lin_svc, rbf_svc, poly_svc)):
    # Plot the decision boundary. For that, we will assign a color to each
    # point in the mesh [x_min, m_max]x[y_min, y_max].
    plt.figure(figsize=(15,15))
    # Set the subplot position (Size = 2 by 2, position deifined by i count
    plt.subplot(2, 2, i + 1)
    
    #Subplot spacing
    plt.subplots_adjust(wspace=0.4, hspace=0.4)
    
    #define Z sebagai prediction
    #np.c_ digunakan untuk formating agar sesuai dan bisa di-predict (?)
    Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
    
    #put the result into color plot
    Z = Z.reshape(xx.shape)
    
    #Contour plot
    #contourf digunakan untuk fill the contour
    plt.contourf(xx, yy, Z, cmap=plt.cm.terrain, alpha=0.5, linewidhts=0.5)
    
    #Plot the training plots
    plt.scatter(X[:, 0], X[:, 1], c=Y, cmap=plt.cm.Dark2)
    
    #labels and titles
    plt.xlabel('Sepal length')
    plt.ylabel('Sepal width')
    plt.xlim(xx.min(), xx.max())
    plt.ylim(yy.min(), yy.max())
    plt.xticks()
    plt.yticks()
    plt.title(titles[i])
    
plt.show()
```
