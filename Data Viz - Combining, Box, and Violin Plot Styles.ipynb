## Box and Violin Plot

```python
# The normal imports
import numpy as np
from numpy.random import randn
import pandas as pd

# Import the stats library from numpy
from scipy import stats

# These are the plotting modules adn libraries we'll use:
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

# Command so that plots appear in the iPython Notebook
%matplotlib inline
```

```python
#merupakan penjelasan mengenai box plot
url = 'http://en.wikipedia.org/wiki/Box_plot#mediaviewer/File:Boxplot_vs_PDF.svg'
```

```python
data1 = randn(100)

#apa kegunaan ditambah 2?
data2 = randn(100)
```

```python
#harus ada parameter data didalamnya
#hal ini berbeda dengan Python 2
sns.boxplot(data=[data1,data2])
```

```python
#perhatikan ada outliers di dalam boxplot tersebut
#jika ingin outliers di dalam (whiskers) caranya adalah
sns.boxplot(data=[data1,data2],whis=np.inf)
```

```python
#jika dilihat di atas, maka whiskers (atau garis) tersebut memanjang hingga sampai ke tempat outliers itu berada 
```

```python
#membuat boxplot secara horizontal
#anehnya jika kita tidak membuat parameter data, maka boxplot tersebut akan horizontal
sns.boxplot(data1)
```

```python
#vertical
sns.boxplot(data=[data1])
```

## Violin Plots

```python
#Membuat normal distribution
data1 = stats.norm(0,5).rvs(100)

#Dua gamma distribution yang di concat
data2 = np.concatenate([stats.gamma(5).rvs(50)-1,
                       -1*stats.gamma(5).rvs(50)])

#dengan menggunakan boxplot tidak begitu kelihatan perbedaan antara boxplot yang satu dengan boxplot yang lainnya
sns.boxplot(data=[data1,data2], whis=np.inf)
```

```python
#denga menggunakan violin plot jauh lebih kelihatan perbedaannya
#yang kiri sudah jelas merupakan distribusi normal, sedangkan gamma ada terlihat yang mana yang negatif dan mana positif
sns.violinplot(data=[data1,data2])
```

```python
#terlihat sama saja, meskipun sama-sama beda juga
#cuma mungkin violinplot bisa menjadi alternatif untuk digunakan jika boxplot tidak ditemukan perbedaan
```

```python
#kita bisa mengganti bandwith dari kernel untuk density (?)
sns.violinplot(data=[data2],bw=0.01)
```

```python
#basically memasukkan rugplot (semua data point) ke dalam violinplot
sns.violinplot(data=[data1], inner='stick')
```
